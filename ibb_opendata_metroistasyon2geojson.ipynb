```python
import requests
import folium
import urllib
from pandas.io.json import json_normalize
import pandas as pd
import folium
import geopandas
from geopandas import GeoDataFrame
from shapely.geometry import Point
from pyproj import Proj
import numpy as np
```

```python
metro = "https://api.ibb.gov.tr/MetroIstanbul/api/MetroMobile/V2/GetStations"

data = requests.get(metro, verify=False)

type(data.json())
```

```python
df = pd.DataFrame(data.json())
df = pd.DataFrame(df['Data'].tolist(), index=df.index)
df
```

```python
df2 = pd.DataFrame(df['DetailInfo'].tolist(), index=df.index)
df2
```

```python
df1 = df.merge(df2, left_index=True, right_index=True, how='inner')
df1 = df1.drop(['DetailInfo'], axis=1)
df1
```

```python
df1.to_csv('metro_istanbul_istasyon.csv', index=False)
```

```python
df = df1
df
```

```python
#koordinatlarda eksik veri var mı
df['Latitude'].isna().any()
```

```python
#eksik verileri çıkar
df = df.dropna(subset=['Latitude', 'Longitude'])
```

```python
df['Latitude']=df['Latitude'].astype(np.float64)
df['Longitude']=df['Longitude'].astype(np.float64)
```

Lat, Lon koordinatlar karisik. M7 hairicindeki istasyonlar yanlis!

```python
geometry = [Point(xy) for xy in zip(df['Latitude'], df['Longitude'])]
gdf = GeoDataFrame(df, crs=4326, geometry=geometry)
gdf.head()
```

```python
m = folium.Map(location=[41, 29], zoom_start=10)
folium.GeoJson(gdf).add_to(m)
m
```

```python
#dump as file, if you want to save it in file
import geojson
with open("./hastane.geojson",mode="w") as f:
    geojson.dump(gdf,f)
```

```python
# Get the latitude and longitude from result as a list
locations = gdf[['Latitude', 'Longitude']]
coords = locations.values.tolist()

from folium.plugins import MarkerCluster

m = folium.Map([38.5, 27], zoom_start= 8, tiles='cartodbpositron')

marker_cluster = MarkerCluster().add_to(m)

for point in range(0, len(coords)):
    folium.Marker(location = coords[point], popup= 'Ad: ' + str(df['ADI'][point])).add_to(marker_cluster)

m
```
